#Exploratory Data Analysis 

```python
df.columns
```

# **Gender**


```python
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
```

```python
 sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
 sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

From above graph we can see that most of the buyers are females and even the purchasing power of females are greater then man

# **AGE**


```python
df.columns 
```

```python
ax=sns.countplot(data=df, x="Age Group",hue='Gender')
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
#Total Amount vs Age Group 
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=sales_age)
```

from above graphs we can see that most of the buyers are of age group between 26-35 yrs females

# **State**

```python
df.columns
```

```python
#Total number of orders of top 10 states 
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x="State",y="Orders",data=sales_state)
```

```python
#Total amount/sales from top 10 states 
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x="State",y="Amount",data=sales_state)
```

from above graph we can see that unexpectedlly most of the orders are form utterpradesh, maharashtra and karnataka respectivley but total sales/amount is from UP, Karnataka and then Maharashtra.

# **Marital Status**

```python
ax=sns.countplot(data=df, x="Marital_Status")
sns.set(rc={'figure.figsize':(5,5)})
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
#Amount on the basis of marital status
marital_status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x="Marital_Status",y="Amount",hue='Gender',data=marital_status)
```

From above graph we can see that most of the buyers are married(Women) and they have high purchasing power.

# **Occupation**

```python
ax=sns.countplot(data=df, x="Occupation")
sns.set(rc={'figure.figsize':(30,5)})
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
#Total amount/sales from top 10 Occupation
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x="Occupation",y="Amount",data=sales_occupation)
```

from above graph we can see that most of the buyers are working in IT, Healhtcare and Aviation Sector

```python
df.columns
```

# **Product Category**

```python
ax=sns.countplot(data=df, x="Product_Category")
sns.set(rc={'figure.figsize':(30,5)})
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
#Total amount/sales from top 10 Occupation
sales_proCat=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x="Product_Category",y="Amount",data=sales_proCat)
```

From above graph we can see that most of the sold products are form Food, Clothing and Electronics category

# **Top selling Products**

```python
#Total amount/sales from top 10 Occupation
sales_proCat=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x="Product_ID",y="Orders",data=sales_proCat)
```

# **Conclusion**

Married women age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products form food,clothing and electronics category.
